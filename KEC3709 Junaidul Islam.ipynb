```python
import random
import string

def display_menu():
    print("\n===== MENU PROGRAM =====")
    print("1. Generate Password")
    print("2. Reverse String")
    print("3. Write a Few Lines to File ('output.txt')")
    print("4. Read and Display File Content ('output.txt')")
    print("5. Exit")
    print("========================")

def generate_password():
    length = int(input("Enter the password length: "))
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    print(f"Generated Password: {password}")

def reverse_string():
    user_string = input("Enter a string to reverse: ")
    reversed_string = user_string[::-1]
    print(f"Reversed String: {reversed_string}")

def write_to_file():
    lines = []
    print("Enter lines of text (type 'END' to stop):")
    while True:
        line = input()
        if line.upper() == "END":
            break
        lines.append(line)

    with open("output.txt", "w") as file:
        for line in lines:
            file.write(line + "\n")
    print("Lines written to 'output.txt' successfully.")

def read_from_file():
    try:
        with open("output.txt", "r") as file:
            print("\nContent of 'output.txt':")
            print(file.read())
    except FileNotFoundError:
        print("File 'output.txt' does not exist. Please write to the file first.")

def main():
    while True:
        display_menu()
        choice = input("Select an option (1-5): ")

        if choice == '1':
            generate_password()
        elif choice == '2':
            reverse_string()
        elif choice == '3':
            write_to_file()
        elif choice == '4':
            read_from_file()
        elif choice == '5':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")
if __name__ == "__main__":
    main()
```
